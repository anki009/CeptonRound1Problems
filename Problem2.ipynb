```python
def solution2():
  a = input()
  a= list(a)
  c=[]
  str=''
  b=['+','-','*','/']
  for i in b:
    for j in b:
      for k in b:
      
        a[1]=i
        a[3]=j
        a[5]=k

        for m in a:
          str += m
        
        c.append(str)
        str=''

  for i in c:
    b= eval(i)
    print (" {} = {} "
       .format(i, b))
```
